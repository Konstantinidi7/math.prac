Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('https://lectures.scientific-python.org/_downloads/087d8fe936fcb7a4762b9850b932cf9e/populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

plt.figure(figsize=(12, 6))
plt.plot(years, hares, label='Зайцы', color='royalblue', linewidth=3, linestyle='--', marker='o', markersize=8)
plt.plot(years, lynxes, label='Рысь', color='orange', linewidth=3, linestyle='-', marker='s', markersize=8)
plt.plot(years, carrots, label='Морковь', color='forestgreen', linewidth=3, linestyle=':', marker='D', markersize=8)

plt.title('Популяциия зайцев, рыси и морковок', fontsize=14)
plt.xlabel('Год', fontsize=14)
plt.ylabel('Размер популяции', fontsize=14)

plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
````

Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
import pandas as pd


data = {
    'year': [1900, 1901, 1902, 1903, 1904, 1905, 1906, 1907, 1908, 1909,
             1910, 1911, 1912, 1913, 1914, 1915, 1916, 1917, 1918, 1919],
    'hare': [30e3, 47.283e3, 70.2e3, 77.4e3, 36.3e3, 20.623e3, 18.1e3,
             21.4e3, 22e3, 25.4e3, 27.1e3, 40.383e3, 57e3, 76.6e3, 52.3e3,
             19.5e3, 11.2e3, 7.623e3, 14.6e3, 16.2e3],
    'carrot': [48300, 48200, 41500, 38200, 40600, 39800, 38600, 42300,
               44500, 42100, 46000, 46800, 43800, 48900, 39400, 39000,
               36700, 41800, 43300, 41300]
}


df = pd.DataFrame(data)

correlation = df['hare'].corr(df['carrot'])
print(f"Коэффициент корреляции между популяцией зайцев и морковок: {correlation}")


carrots_1920 = 47300


mean_hare = df['hare'].mean()
mean_carrot = df['carrot'].mean()


predicted_hare = mean_hare + (correlation * (carrots_1920 - mean_carrot))
hare_accuracy = abs(predicted_hare- 24.7e3)
print(f"Прогнозируемая популяция зайцев в 1920 году: {predicted_hare}")
print(f"Разница с истинным значением для популяции зайцев в 1920 году: {hare_accuracy}")


````

Задание:

1) напишите функцию, реализующую алгоритм k-means.

Входом в функцию является

* массив NxM, где N -- число объектов, M -- число признаков объекта
* k -- число кластеров
* max_iter -- максимальное число итераций
* threshold -- критерий остановки (если на очередной итерации метода суммарное квадратичное отклонение изменяется меньше, чем на threshold, итерации прекращаются)

Выход:

* массив длины N, содержащий индексы кластеров от 0 до k-1
* суммарное квадратичное отклонение для полученного разбиения на кластеры

Постарайтесь обойтись минимальным количеством циклов в вашем коде.

2) Выделите 10 кластеров в датасете MNIST. Дайте интерпретацию каждому полученному кластеру (к какой цифре его можно отнести). Нарисуйте изображения, которые отвечают центрам кластеров. Посчитайте, сколько из точек кластера действительно являются той цифрой, к которой вы его отнесли (для этого сравните с массивом правильных ответов labels) и нарисуйте матрицу ошибок.

````python
import numpy as np

def kmeans(X, k, max_iter=100, threshold=1e-4):
    N, M = X.shape
    centroids = X[np.random.choice(N, k, replace=False)]

    prev_sse = np.inf
    for i in range(max_iter):
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        clusters = np.argmin(distances, axis=1)
        new_centroids = np.array([X[clusters == j].mean(axis=0) for j in range(k)])

        sse = np.sum((X - centroids[clusters]) ** 2)
        if np.abs(prev_sse - sse) < threshold:
            break
        prev_sse = sse
        centroids = new_centroids

    return clusters, sse
````

````python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np

images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
X = images / 255.0

k = 10
clusters, sse = kmeans(X, k)

def plot_cluster_centers(centroids, image_shape=(28, 28)):
    fig, axes = plt.subplots(1, 10, figsize=(10, 2))
    for i, centroid in enumerate(centroids):
        axes[i].imshow(centroid.reshape(image_shape), cmap='gray')
        axes[i].axis('off')
    plt.show()

centroids = np.array([X[clusters == i].mean(axis=0) for i in range(k)])
plot_cluster_centers(centroids)

y_pred = np.zeros_like(labels)
for i in range(k):
    mask = (clusters == i)
    if np.any(mask):
        y_pred[mask] = np.bincount(labels[mask].astype(int)).argmax()

y_pred = y_pred.astype(int)
labels = labels.astype(int)

conf_matrix = confusion_matrix(labels, y_pred)

plt.matshow(conf_matrix, cmap='coolwarm')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
````
